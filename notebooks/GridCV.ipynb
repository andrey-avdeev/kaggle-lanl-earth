```python
import pickle
MODELS_DIR='cache/models/'
FAST_MODEL_DIR=MODELS_DIR+'fast/'
SLOW_MODEL_DIR=MODELS_DIR+'slow/'
FAST_SLOW_MODEL_DIR=MODELS_DIR+'fast_slow/'

def save_fast(best_params, best_score, time_left, name):
    result={
    'best_params':best_params,
    'best_score':best_score,
    'time_left':time_left,
}
    
    with open(FAST_MODEL_DIR+name+'.pkl', 'wb') as fobj:
        pickle.dump(result, fobj)
```
